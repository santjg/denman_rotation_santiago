### NIDAQ Data
There are 8 digital lines and 2 analog lines. Load the `.bin` file into the notebook. From the `.bin`, separate it into np arrays for each analog and digital channel. 

```python
def get_chunk(mm,start,end,channels,sampling_rate=30000,remove_offset=False):
	chunk = mm[int(start*sampling_rate*int(channels)):int(np.floor(end*sampling_rate*(int(channels))))]
	chunk = np.reshape(chunk,(int(channels),-1),order='F')  * 0.195
	# print(np.shape(chunk))
	if remove_offset:
		for i in np.arange(channels):
			chunk[i,:] = chunk[i,:] - np.mean(chunk[i,:])
	return chunk
```

```python
def cross_from_below(x, threshold):
    """
    return the indices into *x* where *x* crosses some threshold from below.
    """
    x = np.asarray(x)
    ind = np.nonzero((x[:-1] < threshold) & (x[1:] >= threshold))[0]
    if len(ind):
        return ind+1
    else:
        return ind
```

```python
nidq_mm = np.memmap(nidq_bin)

XA0 = get_chunk(nidq_mm, 0, 1000, 1)
XA1 = get_chunk(nidq_mm, 0, 1000, 2)
XA2 = get_chunk(nidq_mm, 0, 1000, 3)
XA3 = get_chunk(nidq_mm, 0, 1000, 4)
XA4 = get_chunk(nidq_mm, 0, 1000, 5)
XA5 = get_chunk(nidq_mm, 0, 1000, 6)
XA6 = get_chunk(nidq_mm, 0, 1000, 7)
XA7 = get_chunk(nidq_mm, 0, 1000, 8)
XDO = get_chunk(nidq_mm, 0, 1000, 9)
```

```python
#Get sample numbers for each channel

sample_num_XA0 = cross_from_below(XA0, 1)
sample_num_XA1 = cross_from_below(XA1, 1)
sample_num_XA2 = cross_from_below(XA2, 1)
sample_num_XA3 = cross_from_below(XA3, 1)
sample_num_XA4 = cross_from_below(XA4, 1)
sample_num_XA5 = cross_from_below(XA5, 1)
sample_num_XA6 = cross_from_below(XA6, 1)
sample_num_XA7 = cross_from_below(XA7, 1)
sample_num_XDO = cross_from_below(XDO, 1)
```
